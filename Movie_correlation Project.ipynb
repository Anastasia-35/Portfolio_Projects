```python
#import libraries
import pandas as pd
import numpy as np

#import csv file
data = pd.read_csv("movies.csv")
data.head() #display the first few rows

```

```python

# Checking for missing data
for col in data.columns:
    missing = np.mean(data[col].isnull()) * 100  
    print('{} - {:.2f}%'.format(col, missing))
```

```python
# looking data types for the columns
data.dtypes
```

```python
# create a correct year column

data['monthcorrect'] = data['released'].astype(str).str[:4]
data
```

```python
data.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#drop uplicates

data ['company'].drop_duplicates().sort_values(ascending=True)
```

```python

#scatter plot comparing bugdet vs revenue
import matplotlib.pyplot as plt

plt.scatter(data['budget'], data['gross'])
plt.title('Budget VS Gross Earnings')
plt.xlabel('bugdet')
plt.ylabel('gross')
plt.grid()
plt.show()
```

```python
#plotting budget vs gross using seaborn
import seaborn as sns

sns.regplot(x='budget', y='gross', data=data, scatter_kws={'color' : 'pink'}, line_kws={'color':'purple'})
```

```python
# Select only numeric columns
numeric_data = data.select_dtypes(include=['number'])

# Calculate the correlation matrix for numeric columns
correlation_matrix = numeric_data.corr(method='pearson')

# Display the correlation matrix
correlation_matrix
```

```python
#visualizing using a heatMap

#  Calculate the correlation matrix
correlation_matrix = numeric_data.corr(method='pearson')

#  Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)

# Display the heatmap
plt.title('Correlation Heatmap')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features ')
plt.show()


```

```python
#
df_numerized = data

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
#visualizing using a heatMap
import seaborn as sns
import matplotlib.pyplot as plt

#  Calculate the correlation matrix
correlation_matrix = df_numerized.corr(method='pearson')

#  Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)

# Display the heatmap
plt.title('Correlation Heatmap')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features ')
plt.show()
```

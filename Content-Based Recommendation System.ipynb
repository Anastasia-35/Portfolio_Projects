```python
#Load and prepare the datase
import pandas as pd

# Load the Netflix dataset
df = pd.read_csv('netflix_titles.csv')

#display the data
df
```

```python
# Preprocess the Data

# Drop rows with missing values in important columns
df = df.dropna(subset=['title', 'description', 'cast'])

# Drop unnecessary columns
df = df.drop(columns=['show_id'])

# Convert descriptions to lowercase for consistency
df['description'] = df['description'].apply(lambda x: x.lower())
```

```python
# feature enginnering

# Combine description, cast, and director into one feature
df['combined_features'] = df['description'] + ' ' + df['cast'] + ' ' + df['director']
```

```python
# Normalize the titles in the dataset (strip spaces, lowercase)
df.loc[:, 'normalized_title'] = df['title'].str.strip().str.lower()

# Check the first few rows to confirm the normalization
print(df[['title', 'normalized_title']].head())

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Create a TfidfVectorizer instance
tfidf_vectorizer = TfidfVectorizer(stop_words='english')

# Fit the TF-IDF vectorizer on the normalized titles and transform it into a matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(df['normalized_title'])

# Compute the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Check the shape of the cosine similarity matrix
print(cosine_sim.shape)  
```

```python
#Recommendation function
def recommend_show(title, cosine_sim=cosine_sim, top_n=10):
    # Normalize the title search (strip spaces, lowercase)
    title = title.lower().strip()  # Normalize the input title

    # Check if the title exists in the normalized dataset
    if title not in df['normalized_title'].values:
        return f"'{title}' not found in the dataset."

    # Get the index of the show
    idx = df[df['normalized_title'] == title].index[0]

    # Get the pairwise similarity scores of all shows with that show
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the shows based on the similarity score
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the indices of the top N most similar shows
    sim_scores = sim_scores[1:top_n+1]  # Skip the first item (it will be the same show)
    show_indices = [i[0] for i in sim_scores]

    # Return the top N most similar shows
    return df['title'].iloc[show_indices]


```

```python
# Test the recommendation function with the title 'Ganglands'
recommended_shows = recommend_show('ganglands', top_n=5)

# Print the recommended shows
print("Recommended Shows for 'Ganglands':")
print(recommended_shows)
```
